```python
data = pd.merge(df, rdf, on='movieId')
```

```python
user_rating_pt = data.pivot_table(index='userId', columns='title', values='rating')
```

```python
user_rating_pt.head()
```

```python
toy_story_ratings = user_rating_pt['Toy Story (1995)']
```

```python
similar_movies = user_rating_pt.corrwith(toy_story_ratings)
print(similar_movies)
print("=============")
corr_toy_story = pd.DataFrame(similar_movies, columns=['Correlation'])
print(corr_toy_story)
print("=============")
corr_toy_story.sort_values('Correlation', ascending=False, inplace=True)
print(corr_toy_story)
print("=============")
```

```python
corr_toy_story = corr_toy_story.join(data.groupby('title')['rating'].count())
corr_toy_story.rename(columns={'rating': 'rating_count'}, inplace=True)
```

```python
mask = corr_toy_story['rating_count'] > 100
recommendations = corr_toy_story[mask]
recommendations.sort_values('Correlation', ascending=False)
recommendations.iloc[0:4].head()
```
